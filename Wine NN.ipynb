```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
%matplotlib inline
```

```python
#Activation
def sigmoid(x):
    return 1/ (1 + np.exp(-x))

#Transfer derivative
def dsigmoid(x):
    return x * (1-x)
```

```python
wine = load_wine()

features = wine.data
target = wine.target
names = wine.feature_names
```

```python
def feed_forward(x, w0, w1, w2):
    #input layer
    layer0 = x
    layer1 = sigmoid(np.dot(layer0, w0))
    layer2 = sigmoid(np.dot(layer1, w1))
    layer3 = sigmoid(np.dot(layer2, w2))
    
    return layer0, layer1, layer2, layer3
```

```python
def backprop(i, a0, a1, a2, a3, w0, w1, w2, y, lr):
    
    l3_error = a3 - y
    l3_delta = l3_error * dsigmoid(a3)
    dh3 = np.dot(a2.T, l3_delta)
    
    l2_error = l3_delta.dot(w2.T)
    l2_delta = l2_error * dsigmoid(a2)
    dh2 = np.dot(a1.T, l2_delta)
    
    l1_error = l2_delta.dot(w1.T)
    l1_delta = l1_error * dsigmoid(a1)
    dh1 = np.dot(a0.T, l1_delta)
    
    w2 = w2 - (dh3 * lr)
    w1 = w1 - (dh2 * lr)
    w0 = w0 - (dh1 * lr)
    
    loss = np.mean(np.power(l3_error, 2))
    loss_curve.append(loss)
    
    if i%1000 == 0:
        print(i, "Error:", loss)
        
    return w0, w1, w2
```

```python
w0 = np.random.random((13,16))
w1 = np.random.random((16,10))
w2 = np.random.random((10,3))
```

```python
iterations = range(20000)
loss_curve = []
lr = 0.01
```

```python
for j in iterations:
    a0, a1, a2, a3 = feed_forward(features,w0,w1,w2)
    w0,w1,w2 = backprop(j, a0, a1, a2, a3, w0, w1, w2, target, lr)
```

```python
def accuracy(testx, testy):
    correct = 0
    a0, a1, a2, a3 = feed_forward(testx, w0, w1, w2)
    
    for i in range(len(testx)):
        if np.argmax(a3[i]) == np.argmax(testy[i]):
            correct += 1
    
    return (correct * 100) / len(testx)
```
